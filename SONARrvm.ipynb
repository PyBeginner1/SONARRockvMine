```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
data = pd.read_csv('/content/Sonar data.csv', header =None)   #header is none since we dont have col names
```

```python
#statistical information of the dataset
data.describe()
```

```python
#groups rock & mines
data[60].value_counts()
```

```python
data.groupby(60).mean()
```

Seperating data & labels

```python
X = data.drop([60], axis =1)
Y =data[60]
```

```python
print(X)
```

```python
print(Y)
```

Train & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.1, random_state =1, stratify = Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

Training Data

```python
#train model with training data
model.fit(X_train, Y_train)
```

```python
Model Evauluation
```

```python
#accuracy training data
predict_training_data = model.predict(X_train)
```

```python
score = accuracy_score(Y_train,predict_training_data )
print("Accuracy score is " ,score)
```

```python
#accuracy on test data
predict_test_data = model.predict(X_test)
```

```python
score_1 = accuracy_score(predict_test_data, Y_test)
print("Accuracy of testing data is " , score_1)
```
